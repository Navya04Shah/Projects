```python
#Understanding axes, scales, and legends

import matplotlib.pyplot as plt

# Example: Population growth chart
years = [2010, 2011, 2012, 2013, 2014]
population = [112, 115, 118, 122, 125]  # in millions

plt.figure(figsize=(8, 5))
plt.plot(years, population, marker='o', linestyle='-', color='b')
plt.xlabel('Year')
plt.ylabel('Population (millions)')
plt.title('Population Growth Over Years')
plt.grid(True)
plt.show()
```

```python
# Example: Scatter plot showing correlation between study hours and exam scores
import numpy as np

np.random.seed(0)
study_hours = np.random.randint(1, 10, 20)  # random study hours
exam_scores = study_hours * 10 + np.random.randint(-5, 5, 20)  # simulated exam scores

plt.figure(figsize=(8, 5))
plt.scatter(study_hours, exam_scores, color='g')
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
plt.title('Relationship between Study Hours and Exam Scores')
plt.grid(True)
plt.show()
```

```python
# Example: Line chart showing temperature trends over years
years = range(2010, 2020)
temperatures = [20, 21, 23, 22, 25, 24, 26, 27, 28, 29]  # hypothetical temperatures

plt.figure(figsize=(8, 5))
plt.plot(years, temperatures, marker='o', linestyle='-', color='r')
plt.xlabel('Year')
plt.ylabel('Temperature (C)')
plt.title('Temperature Trends Over Years')
plt.grid(True)
plt.show()
```

```python
# Example: Pie chart showing market share
categories = ['Category A', 'Category B', 'Category C']
market_shares = [40, 30, 30]  # percentages

plt.figure(figsize=(6, 6))
plt.pie(market_shares, labels=categories, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue'])
plt.title('Market Share Distribution')
plt.show()
```
